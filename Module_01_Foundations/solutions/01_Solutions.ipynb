## Упражнение 1: Бросок двух кубиков

```python
# Упражнение 1: Бросок двух кубиков
print('Упражнение 1: Бросок двух кубиков')
print('=' * 60)

# Все возможные исходы
outcomes = list(product(range(1, 7), repeat=2))
n_outcomes = len(outcomes)

# 1. Вероятность суммы 7
sum_7 = [o for o in outcomes if sum(o) == 7]
P_sum_7 = len(sum_7) / n_outcomes
print(f'\n1. P(сумма = 7):')
print(f'   Исходы с суммой 7: {sum_7}')
print(f'   Количество: {len(sum_7)}')
print(f'   P(сумма = 7) = {len(sum_7)}/{n_outcomes} = {P_sum_7:.4f}')

# 2. P(сумма = 7 | первый кубик = 3)
first_is_3 = [o for o in outcomes if o[0] == 3]
sum_7_first_3 = [o for o in first_is_3 if sum(o) == 7]
P_sum_7_given_first_3 = len(sum_7_first_3) / len(first_is_3)

print(f'\n2. P(сумма = 7 | первый = 3):')
print(f'   Исходы, где первый = 3: {first_is_3}')
print(f'   Из них с суммой 7: {sum_7_first_3}')
print(f'   P(сумма = 7 | первый = 3) = {len(sum_7_first_3)}/{len(first_is_3)} = {P_sum_7_given_first_3:.4f}')

# 3. Независимость
P_first_3 = len(first_is_3) / n_outcomes
P_independent = P_first_3 * P_sum_7

print(f'\n3. Независимость событий:')
print(f'   P(первый = 3) = {P_first_3:.4f}')
print(f'   P(сумма = 7) = {P_sum_7:.4f}')
print(f'   P(первый = 3) * P(сумма = 7) = {P_independent:.4f}')
print(f'   P(первый = 3 и сумма = 7) = {len(sum_7_first_3)/n_outcomes:.4f}')
print(f'   Независимы? {np.isclose(P_independent, len(sum_7_first_3)/n_outcomes)}')
```

## Упражнение 2: Диагностика болезни

```python
# Упражнение 2: Диагностика болезни
print('\nУпражнение 2: Диагностика болезни')
print('=' * 60)

# Параметры
P_disease = 0.02
P_healthy = 0.98
P_pos_given_disease = 0.90
P_neg_given_healthy = 0.95
P_pos_given_healthy = 1 - P_neg_given_healthy

# 1. P(Test+)
P_positive = P_pos_given_disease * P_disease + P_pos_given_healthy * P_healthy
print(f'\n1. P(Test+):')
print(f'   P(Test+) = P(Test+|Disease) * P(Disease) + P(Test+|Healthy) * P(Healthy)')
print(f'   P(Test+) = {P_pos_given_disease} * {P_disease} + {P_pos_given_healthy} * {P_healthy}')
print(f'   P(Test+) = {P_positive:.4f}')

# 2. P(Disease|Test+)
P_disease_given_pos = (P_pos_given_disease * P_disease) / P_positive
print(f'\n2. P(Disease|Test+):')
print(f'   P(Disease|Test+) = P(Test+|Disease) * P(Disease) / P(Test+)')
print(f'   P(Disease|Test+) = {P_pos_given_disease} * {P_disease} / {P_positive:.4f}')
print(f'   P(Disease|Test+) = {P_disease_given_pos:.4f}')

# 3. Изменение при заболеваемости 20%
P_disease_new = 0.20
P_healthy_new = 0.80
P_positive_new = P_pos_given_disease * P_disease_new + P_pos_given_healthy * P_healthy_new
P_disease_given_pos_new = (P_pos_given_disease * P_disease_new) / P_positive_new

print(f'\n3. При заболеваемости 20%:')
print(f'   P(Test+) = {P_positive_new:.4f}')
print(f'   P(Disease|Test+) = {P_disease_given_pos_new:.4f}')
print(f'\n   Вывод: При более высокой заболеваемости')
print(f'   вероятность заболевания при положительном тесте увеличивается.')
```

## Упражнение 3: Урна с шарами

```python
# Упражнение 3: Урна с шарами
print('\nУпражнение 3: Урна с шарами')
print('=' * 60)

from itertools import combinations

# Урна: 5 белых и 3 чёрных
urn = ['W1', 'W2', 'W3', 'W4', 'W5', 'B1', 'B2', 'B3']

# Все возможные пары
all_pairs = list(combinations(urn, 2))
n_pairs = len(all_pairs)

# 1. P(оба белые)
white_pairs = [p for p in all_pairs if p[0][0] == 'W' and p[1][0] == 'W']
P_both_white = len(white_pairs) / n_pairs

print(f'\n1. P(оба белые):')
print(f'   Всего пар: {n_pairs}')
print(f'   Пар с двумя белыми: {len(white_pairs)}')
print(f'   P(оба белые) = {len(white_pairs)}/{n_pairs} = {P_both_white:.4f}')

# 2. P(второй белый | первый белый)
first_white = [p for p in all_pairs if p[0][0] == 'W']
second_white_first_white = [p for p in first_white if p[1][0] == 'W']
P_second_white_given_first = len(second_white_first_white) / len(first_white)

print(f'\n2. P(второй белый | первый белый):')
print(f'   Пары, где первый белый: {len(first_white)}')
print(f'   Из них второй тоже белый: {len(second_white_first_white)}')
print(f'   P(второй белый | первый белый) = {len(second_white_first_white)}/{len(first_white)} = {P_second_white_given_first:.4f}')
```

## Упражнение 4: Независимость событий

```python
# Упражнение 4: Независимость событий
print('\nУпражнение 4: Независимость событий')
print('=' * 60)

# Случай 1
P_A1, P_B1, P_AB1 = 0.3, 0.4, 0.12
is_independent1 = np.isclose(P_AB1, P_A1 * P_B1)

print(f'\nСлучай 1:')
print(f'  P(A) = {P_A1}, P(B) = {P_B1}, P(A∩B) = {P_AB1}')
print(f'  P(A) * P(B) = {P_A1 * P_B1}')
print(f'  Независимы? {is_independent1}')

# Случай 2
P_A2, P_B2, P_AB2 = 0.5, 0.5, 0.3
is_independent2 = np.isclose(P_AB2, P_A2 * P_B2)

print(f'\nСлучай 2:')
print(f'  P(A) = {P_A2}, P(B) = {P_B2}, P(A∩B) = {P_AB2}')
print(f'  P(A) * P(B) = {P_A2 * P_B2}')
print(f'  Независимы? {is_independent2}')
```
