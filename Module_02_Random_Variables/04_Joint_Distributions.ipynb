## 1. Совместная функция распределения

**Совместная функция распределения** двух случайных величин $X$ и $Y$:

$$F_{XY}(x, y) = P(X \leq x, Y \leq y)$$

### Для дискретных случайных величин:
**Совместная функция вероятности (PMF):**
$$p_{XY}(x, y) = P(X = x, Y = y)$$

Свойства:
- $p_{XY}(x, y) \geq 0$
- $\sum_{x} \sum_{y} p_{XY}(x, y) = 1$

### Для непрерывных случайных величин:
**Совместная плотность (PDF):**
$$f_{XY}(x, y) \geq 0$$
$$\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f_{XY}(x, y) \, dx \, dy = 1$$

$$P((X, Y) \in A) = \int\int_A f_{XY}(x, y) \, dx \, dy$$

```python
# Пример: Совместное распределение дискретных случайных величин
# Бросок двух кубиков: X = первый кубик, Y = второй кубик

# Совместная PMF (таблица)
joint_pmf = np.zeros((6, 6))
for i in range(6):
    for j in range(6):
        joint_pmf[i, j] = 1/36

# Визуализация как тепловая карта
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Тепловая карта
im = axes[0].imshow(joint_pmf, cmap='YlOrRd', origin='lower')
axes[0].set_xlabel('Y (второй кубик)')
axes[0].set_ylabel('X (первый кубик)')
axes[0].set_title('Совместная PMF двух кубиков')
axes[0].set_xticks(range(6))
axes[0].set_xticklabels(range(1, 7))
axes[0].set_yticks(range(6))
axes[0].set_yticklabels(range(1, 7))
plt.colorbar(im, ax=axes[0], label='P(X=x, Y=y)')

# 3D визуализация
x = np.arange(6)
y = np.arange(6)
X, Y = np.meshgrid(x, y)
axes[1] = fig.add_subplot(122, projection='3d')
axes[1].plot_surface(X, Y, joint_pmf.T, cmap='YlOrRd', alpha=0.8)
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].set_zlabel('P(X,Y)')
axes[1].set_title('3D визуализация совместной PMF')

plt.tight_layout()
plt.show()
```

## 2. Маргинальные распределения

**Маргинальное распределение** — это распределение одной случайной величины без учёта другой.

### Для дискретных случайных величин:
$$p_X(x) = \sum_{y} p_{XY}(x, y)$$
$$p_Y(y) = \sum_{x} p_{XY}(x, y)$$

### Для непрерывных случайных величин:
$$f_X(x) = \int_{-\infty}^{\infty} f_{XY}(x, y) \, dy$$
$$f_Y(y) = \int_{-\infty}^{\infty} f_{XY}(x, y) \, dx$$

**Интуитивно:** Маргинальное распределение получается "суммированием" по одной из переменных.

```python
# Пример: Маргинальные распределения
# Данные: оценки студентов по двум предметам

np.random.seed(42)
n_students = 200

# Генерация коррелированных оценок
math_scores = np.random.normal(70, 10, n_students)
physics_scores = 0.7 * math_scores + np.random.normal(0, 8, n_students)

# Округление до целых
math_scores = np.clip(np.round(math_scores), 0, 100).astype(int)
physics_scores = np.clip(np.round(physics_scores), 0, 100).astype(int)

# Создание DataFrame
df = pd.DataFrame({'Математика': math_scores, 'Физика': physics_scores})

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Совместное распределение (scatter)
axes[0, 0].scatter(df['Математика'], df['Физика'], alpha=0.5, color='steelblue')
axes[0, 0].set_xlabel('Оценка по математике')
axes[0, 0].set_ylabel('Оценка по физике')
axes[0, 0].set_title('Совместное распределение оценок')

# 2D гистограмма
axes[0, 1].hist2d(df['Математика'], df['Физика'], bins=20, cmap='YlOrRd')
axes[0, 1].set_xlabel('Оценка по математике')
axes[0, 1].set_ylabel('Оценка по физике')
axes[0, 1].set_title('2D гистограмма')
plt.colorbar(axes[0, 1].collections[0], ax=axes[0, 1], label='Количество')

# Маргинальное распределение математики
axes[1, 0].hist(df['Математика'], bins=20, density=True, alpha=0.7, color='blue')
axes[1, 0].set_xlabel('Оценка по математике')
axes[1, 0].set_ylabel('Плотность')
axes[1, 0].set_title('Маргинальное распределение (Математика)')

# Маргинальное распределение физики
axes[1, 1].hist(df['Физика'], bins=20, density=True, alpha=0.7, color='red')
axes[1, 1].set_xlabel('Оценка по физике')
axes[1, 1].set_ylabel('Плотность')
axes[1, 1].set_title('Маргинальное распределение (Физика)')

plt.tight_layout()
plt.show()

# Статистики
print('Маргинальные статистики:')
print('=' * 50)
print(f'Математика: E[X] = {df["Математика"].mean():.2f}, σ = {df["Математика"].std():.2f}')
print(f'Физика: E[Y] = {df["Физика"].mean():.2f}, σ = {df["Физика"].std():.2f}')
print(f'Корреляция: r = {df.corr().iloc[0, 1]:.4f}')
```

## 3. Условные распределения

**Условное распределение** $X$ при условии $Y = y$:

### Для дискретных случайных величин:
$$p_{X|Y}(x|y) = \frac{p_{XY}(x, y)}{p_Y(y)}$$

### Для непрерывных случайных величин:
$$f_{X|Y}(x|y) = \frac{f_{XY}(x, y)}{f_Y(y)}$$

**Условное математическое ожидание:**
$$E[X|Y=y] = \sum_{x} x \cdot p_{X|Y}(x|y)$$

**Закон полного математического ожидания (Adam's Law):**
$$E[X] = E[E[X|Y]]$$

```python
# Пример: Условные распределения
# Продолжаем с данными об оценках

# Условное распределение физики при разных уровнях математики
math_levels = [(50, 60), (60, 70), (70, 80), (80, 90)]
colors = ['red', 'blue', 'green', 'purple']

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Условные распределения
for idx, (low, high) in enumerate(math_levels):
    mask = (df['Математика'] >= low) & (df['Математика'] < high)
    physics_given_math = df.loc[mask, 'Физика']
    
    if len(physics_given_math) > 0:
        axes[0].hist(physics_given_math, bins=15, density=True, alpha=0.5, 
                     color=colors[idx], label=f'Math ∈ [{low},{high})')

axes[0].set_xlabel('Оценка по физике')
axes[0].set_ylabel('Плотность')
axes[0].set_title('Условные распределения физики при разных уровнях математики')
axes[0].legend()

# Условные средние
math_bins = np.arange(40, 100, 10)
conditional_means = []
math_centers = []

for i in range(len(math_bins) - 1):
    mask = (df['Математика'] >= math_bins[i]) & (df['Математика'] < math_bins[i+1])
    if mask.sum() > 0:
        conditional_means.append(df.loc[mask, 'Физика'].mean())
        math_centers.append((math_bins[i] + math_bins[i+1]) / 2)

axes[1].scatter(df['Математика'], df['Физика'], alpha=0.3, color='gray', label='Данные')
axes[1].plot(math_centers, conditional_means, 'ro-', linewidth=2, markersize=8, 
             label='E[Physics|Math]')
axes[1].axhline(y=df['Физика'].mean(), color='blue', linestyle='--', 
                label=f'E[Physics] = {df["Физика"].mean():.1f}')
axes[1].set_xlabel('Оценка по математике')
axes[1].set_ylabel('Оценка по физике')
axes[1].set_title('Условное математическое ожидание')
axes[1].legend()

plt.tight_layout()
plt.show()
```

## 4. Независимость случайных величин

**Определение:** Случайные величины $X$ и $Y$ **независимы**, если:

$$F_{XY}(x, y) = F_X(x) \cdot F_Y(y)$$

или эквивалентно:

$$p_{XY}(x, y) = p_X(x) \cdot p_Y(y)$$

или для непрерывных:

$$f_{XY}(x, y) = f_X(x) \cdot f_Y(y)$$

### Свойства независимых случайных величин:
1. $\text{Cov}(X, Y) = 0$ и $\rho_{XY} = 0$ (обратное неверно!)
2. $E[XY] = E[X] \cdot E[Y]$
3. $\text{Var}(X + Y) = \text{Var}(X) + \text{Var}(Y)$
4. $E[X|Y] = E[X]$ (знание $Y$ не помогает предсказать $X$)

**Важно:** $\rho = 0$ не означает независимость! (только для нормального распределения)

```python
# Пример: Независимость vs нулевая корреляция
np.random.seed(42)
n = 1000

# Случайные величины с нулевой корреляцией, но НЕ независимые
X = np.random.uniform(-1, 1, n)
Y = X**2 + np.random.normal(0, 0.1, n)  # Y зависит от X^2

r = np.corrcoef(X, Y)[0, 1]

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Независимые случайные величины
X1 = np.random.normal(0, 1, n)
Y1 = np.random.normal(0, 1, n)
r1 = np.corrcoef(X1, Y1)[0, 1]

axes[0].scatter(X1, Y1, alpha=0.3, color='blue')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title(f'Независимые\nr = {r1:.3f}')

# Коррелированные, но не независимые
X2 = np.random.normal(0, 1, n)
Y2 = 0.8 * X2 + np.random.normal(0, 0.6, n)
r2 = np.corrcoef(X2, Y2)[0, 1]

axes[1].scatter(X2, Y2, alpha=0.3, color='green')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].set_title(f'Коррелированные\nr = {r2:.3f}')

# Нулевая корреляция, но зависимые!
axes[2].scatter(X, Y, alpha=0.3, color='red')
axes[2].set_xlabel('X')
axes[2].set_ylabel('Y')
axes[2].set_title(f'Зависимые, но r = 0\nr = {r:.3f}')

plt.tight_layout()
plt.show()

print('Пример: Нулевая корреляция ≠ Независимость')
print('=' * 50)
print(f'Корреляция между X и Y = X²: r = {r:.4f}')
print(f'\nХотя корреляция ≈ 0, Y полностью определяется X!')
print(f'\nВывод: Корреляция измеряет только линейную связь.')
print(f'Независимость — более сильное условие.')
```

## 5. Многомерное нормальное распределение

**Мномерное нормальное распределение** — это обобщение нормального распределения на несколько переменных.

$$\mathbf{X} \sim N(\boldsymbol{\mu}, \boldsymbol{\Sigma})$$

где:
- $\boldsymbol{\mu}$ — вектор средних
- $\boldsymbol{\Sigma}$ — ковариационная матрица

**PDF:**
$$f(\mathbf{x}) = \frac{1}{(2\pi)^{k/2} |\boldsymbol{\Sigma}|^{1/2}} \exp\left(-\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{x} - \boldsymbol{\mu})\right)$$

**Свойства:**
- Маргинальные распределения — нормальные
- Условные распределения — нормальные
- Линейные комбинации — нормальные

```python
# Пример: Многомерное нормальное распределение
np.random.seed(42)

# Параметры
mu = np.array([0, 0])  # Вектор средних
Sigma = np.array([[1, 0.8],   # Ковариационная матрица
                  [0.8, 1]])

# Генерация выборки
n_samples = 1000
X = np.random.multivariate_normal(mu, Sigma, n_samples)

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Scatter plot
axes[0].scatter(X[:, 0], X[:, 1], alpha=0.3, color='steelblue')
axes[0].set_xlabel('X₁')
axes[0].set_ylabel('X₂')
axes[0].set_title(f'Многомерное нормальное\nρ = {Sigma[0, 1]}')
axes[0].set_aspect('equal')

# 2D гистограмма
axes[1].hist2d(X[:, 0], X[:, 1], bins=30, cmap='YlOrRd')
axes[1].set_xlabel('X₁')
axes[1].set_ylabel('X₂')
axes[1].set_title('2D гистограмма')
axes[1].set_aspect('equal')
plt.colorbar(axes[1].collections[0], ax=axes[1])

# Маргинальные распределения
axes[2].hist(X[:, 0], bins=30, density=True, alpha=0.5, label='X₁', color='blue')
axes[2].hist(X[:, 1], bins=30, density=True, alpha=0.5, label='X₂', color='red')
x_range = np.linspace(-4, 4, 100)
axes[2].plot(x_range, stats.norm.pdf(x_range), 'b-', linewidth=2)
axes[2].plot(x_range, stats.norm.pdf(x_range), 'r-', linewidth=2)
axes[2].set_xlabel('Значение')
axes[2].set_ylabel('Плотность')
axes[2].set_title('Маргинальные распределения')
axes[2].legend()

plt.tight_layout()
plt.show()

# Ковариационная матрица
print('Ковариационная матрица (эмпирическая):')
print(np.cov(X.T).round(4))
print(f'\nКорреляция: {np.corrcoef(X.T)[0, 1]:.4f}')
```
