## 1. Математическое ожидание

**Математическое ожидание** (期望值) — это среднее значение случайной величины, взвешенное по вероятностям.

### Для дискретной случайной величины:
$$E[X] = \sum_{x} x \cdot P(X = x) = \sum_{x} x \cdot p(x)$$

### Для непрерывной случайной величины:
$$E[X] = \int_{-\infty}^{\infty} x \cdot f(x) dx$$

### Свойства математического ожидания:
1. $E[c] = c$ для константы $c$
2. $E[aX + b] = aE[X] + b$ (линейность)
3. $E[X + Y] = E[X] + E[Y]$ (аддитивность)
4. $E[XY] = E[X] \cdot E[Y]$ для независимых $X$ и $Y$

**Интуитивно:** Математическое ожидание — это "центр тяжести" распределения.

```python
# Пример: Математическое ожидание для дискретного распределения
# Бросок игрального кубика

x = np.array([1, 2, 3, 4, 5, 6])
p = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

# Вычисление E[X]
E_X = np.sum(x * p)
print(f'E[X] = {E_X:.4f}')

# Проверка через numpy (для большого числа экспериментов)
np.random.seed(42)
n_samples = 100000
samples = np.random.choice(x, size=n_samples, p=p)
E_X_empirical = np.mean(samples)
print(f'E[X] (эмпирическое, n={n_samples}) = {E_X_empirical:.4f}')

# Визуализация
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# PMF
axes[0].bar(x, p, color='steelblue', alpha=0.7, edgecolor='black')
axes[0].axvline(x=E_X, color='red', linestyle='--', linewidth=2, label=f'E[X]={E_X:.2f}')
axes[0].set_xlabel('x')
axes[0].set_ylabel('P(X=x)')
axes[0].set_title('Распределение кубика')
axes[0].legend()

# Гистограмма экспериментов
axes[1].hist(samples, bins=np.arange(0.5, 7.5, 1), density=True, 
             alpha=0.7, color='coral', edgecolor='black')
axes[1].axvline(x=E_X, color='red', linestyle='--', linewidth=2, label=f'E[X]={E_X:.2f}')
axes[1].axvline(x=E_X_empirical, color='blue', linestyle='--', linewidth=2, 
                label=f'Среднее={E_X_empirical:.2f}')
axes[1].set_xlabel('x')
axes[1].set_ylabel('Плотность')
axes[1].set_title(f'Эмпирическое распределение (n={n_samples})')
axes[1].legend()

plt.tight_layout()
plt.show()
```

## 2. Дисперсия и стандартное отклонение

**Дисперсия** измеряет разброс случайной величины относительно её среднего.

$$\text{Var}(X) = E[(X - E[X])^2] = E[X^2] - (E[X])^2$$

**Стандартное отклонение:**
$$\sigma = \sqrt{\text{Var}(X)}$$

### Свойства дисперсии:
1. $\text{Var}(c) = 0$ для константы
2. $\text{Var}(aX + b) = a^2 \text{Var}(X)$
3. $\text{Var}(X + Y) = \text{Var}(X) + \text{Var}(Y) + 2\text{Cov}(X,Y)$
4. Для независимых: $\text{Var}(X + Y) = \text{Var}(X) + \text{Var}(Y)$

**Интуитивно:** Стандартное отклонение показывает, насколько в среднем значения отклоняются от среднего.

```python
# Пример: Дисперсия и стандартное отклонение
# Сравнение двух распределений с одинаковым средним, но разным разбросом

np.random.seed(42)

# Два нормальных распределения с одинаковым средним
mu = 50
sigma1, sigma2 = 5, 15

n_samples = 10000
X1 = np.random.normal(mu, sigma1, n_samples)
X2 = np.random.normal(mu, sigma2, n_samples)

print('Сравнение двух распределений:')
print('=' * 50)
print(f'X1 ~ N({mu}, {sigma1}²):')
print(f'  Среднее: {np.mean(X1):.2f}')
print(f'  Дисперсия: {np.var(X1):.2f}')
print(f'  Стандартное отклонение: {np.std(X1):.2f}')
print(f'\nX2 ~ N({mu}, {sigma2}²):')
print(f'  Среднее: {np.mean(X2):.2f}')
print(f'  Дисперсия: {np.var(X2):.2f}')
print(f'  Стандартное отклонение: {np.std(X2):.2f}')

# Визуализация
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Гистограммы
x_range = np.linspace(20, 80, 1000)
axes[0].hist(X1, bins=50, density=True, alpha=0.5, color='blue', label=f'σ={sigma1}')
axes[0].hist(X2, bins=50, density=True, alpha=0.5, color='red', label=f'σ={sigma2}')
axes[0].plot(x_range, stats.norm.pdf(x_range, mu, sigma1), 'b-', linewidth=2)
axes[0].plot(x_range, stats.norm.pdf(x_range, mu, sigma2), 'r-', linewidth=2)
axes[0].axvline(x=mu, color='black', linestyle='--', label=f'μ={mu}')
axes[0].set_xlabel('x')
axes[0].set_ylabel('Плотность')
axes[0].set_title('Два распределения с разной дисперсией')
axes[0].legend()

# Box plot
bp = axes[1].boxplot([X1, X2])
axes[1].set_xticklabels([f'σ={sigma1}', f'σ={sigma2}'])
axes[1].set_ylabel('Значения')
axes[1].set_title('Box plot сравнение')

plt.tight_layout()
plt.show()
```

## 3. Ковариация

**Ковариация** измеряет совместную изменчивость двух случайных величин.

$$\text{Cov}(X, Y) = E[(X - E[X])(Y - E[Y])] = E[XY] - E[X]E[Y]$$

### Свойства:
1. $\text{Cov}(X, X) = \text{Var}(X)$
2. $\text{Cov}(X, Y) = \text{Cov}(Y, X)$ (симметричность)
3. $\text{Cov}(aX + b, cY + d) = ac \cdot \text{Cov}(X, Y)$
4. $\text{Cov}(X + Y, Z) = \text{Cov}(X, Z) + \text{Cov}(Y, Z)$

**Интерпретация:**
- $\text{Cov}(X, Y) > 0$: $X$ и $Y$ возрастают вместе
- $\text{Cov}(X, Y) < 0$: $X$ возрастает, $Y$ убывает
- $\text{Cov}(X, Y) = 0$: нет линейной связи

**Проблема:** Ковариация зависит от масштаба данных!

```python
# Пример: Ковариация
np.random.seed(42)

# Генерация коррелированных данных
n = 100
X = np.random.normal(0, 1, n)
Y_pos = 2 * X + np.random.normal(0, 1, n)  # Положительная связь
Y_neg = -2 * X + np.random.normal(0, 1, n)  # Отрицательная связь
Y_none = np.random.normal(0, 1, n)          # Нет связи

print('Ковариация между X и различными Y:')
print('=' * 50)
print(f'Cov(X, Y_pos) = {np.cov(X, Y_pos)[0,1]:.4f} (положительная)')
print(f'Cov(X, Y_neg) = {np.cov(X, Y_neg)[0,1]:.4f} (отрицательная)')
print(f'Cov(X, Y_none) = {np.cov(X, Y_none)[0,1]:.4f} (нет связи)')

# Визуализация
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

axes[0].scatter(X, Y_pos, alpha=0.5, color='green')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title(f'Положительная ковариация\nCov={np.cov(X, Y_pos)[0,1]:.2f}')

axes[1].scatter(X, Y_neg, alpha=0.5, color='red')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].set_title(f'Отрицательная ковариация\nCov={np.cov(X, Y_neg)[0,1]:.2f}')

axes[2].scatter(X, Y_none, alpha=0.5, color='blue')
axes[2].set_xlabel('X')
axes[2].set_ylabel('Y')
axes[2].set_title(f'Нет линейной связи\nCov={np.cov(X, Y_none)[0,1]:.2f}')

plt.tight_layout()
plt.show()
```

## 4. Корреляция (Коэффициент корреляции Пирсона)

**Корреляция** — это нормированная ковариация, которая принимает значения от -1 до 1.

$$\rho_{XY} = \frac{\text{Cov}(X, Y)}{\sigma_X \cdot \sigma_Y} = \frac{\text{Cov}(X, Y)}{\sqrt{\text{Var}(X) \cdot \text{Var}(Y)}}$$

### Свойства:
1. $-1 \leq \rho \leq 1$
2. $\rho = 1$: идеальная положительная линейная связь
3. $\rho = -1$: идеальная отрицательная линейная связь
4. $\rho = 0$: нет линейной связи (но может быть нелинейная!)

**Важно:** Корреляция не подразумевает причинно-следственную связь!

### Интерпретация значений:
- $|\rho| < 0.3$: слабая корреляция
- $0.3 \leq |\rho| < 0.7$: умеренная корреляция
- $|\rho| \geq 0.7$: сильная корреляция

```python
# Пример: Корреляция
np.random.seed(42)

# Создание данных с разными типами корреляции
n = 200
X = np.random.uniform(-3, 3, n)

datasets = {
    'Сильная положительная': (X, 2*X + np.random.normal(0, 0.5, n)),
    'Слабая положительная': (X, 0.5*X + np.random.normal(0, 2, n)),
    'Нет корреляции': (X, np.random.normal(0, 2, n)),
    'Нелинейная связь': (X, X**2 + np.random.normal(0, 1, n))
}

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for idx, (title, (x, y)) in enumerate(datasets.items()):
    r = np.corrcoef(x, y)[0, 1]
    axes[idx].scatter(x, y, alpha=0.5, color='steelblue')
    axes[idx].set_xlabel('X')
    axes[idx].set_ylabel('Y')
    axes[idx].set_title(f'{title}\nr = {r:.3f}')
    
    # Линия регрессии (только для линейных случаев)
    if 'Нелинейная' not in title:
        z = np.polyfit(x, y, 1)
        p = np.poly1d(z)
        x_line = np.linspace(x.min(), x.max(), 100)
        axes[idx].plot(x_line, p(x_line), 'r-', linewidth=2)

plt.tight_layout()
plt.show()

# Таблица корреляций
print('\nКорреляции:')
print('=' * 40)
for title, (x, y) in datasets.items():
    r = np.corrcoef(x, y)[0, 1]
    print(f'{title}: r = {r:.4f}')
```

## 5. Корреляция vs Причинность

**Критически важно:** Корреляция **не означает** причинно-следственную связь!

### Примеры ложных корреляций:
1. Продажи мороженого и утопления коррелируют (обе связаны с жаркой погодой)
2. Количество пиратов и глобальное потепление (оба тренда во времени)
3. Расходы на рекламу и продажи (может быть обратная причинность)

### Для установления причинности нужны:
- Рандомизированные контролируемые эксперименты (A/B тесты)
- Учёт конфаундеров (confounding variables)
- Временная последовательность (причина предшествует следствию)

```python
# Демонстрация спurious correlation (ложная корреляция)
np.random.seed(42)

# Два случайных блуждания (random walk) могут коррелировать!
n = 100
t = np.arange(n)

# Два независимых случайных блуждания
X_rw = np.cumsum(np.random.normal(0, 1, n))
Y_rw = np.cumsum(np.random.normal(0, 1, n))

r = np.corrcoef(X_rw, Y_rw)[0, 1]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Временные ряды
axes[0].plot(t, X_rw, label='X (random walk)', color='blue')
axes[0].plot(t, Y_rw, label='Y (random walk)', color='red')
axes[0].set_xlabel('Время')
axes[0].set_ylabel('Значение')
axes[0].set_title('Два независимых случайных блуждания')
axes[0].legend()

# Диаграмма рассеяния
axes[1].scatter(X_rw, Y_rw, alpha=0.5, color='green')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].set_title(f'Корреляция между X и Y\nr = {r:.3f}')

plt.tight_layout()
plt.show()

print(f'Корреляция: r = {r:.3f}')
print(f'\nВывод: Даже независимые случайные блуждания')
print(f'могут показывать высокую корреляцию!')
print(f'Это пример ложной (spurious) корреляции.')
```

## 6. Практический пример: Анализ портфеля акций

Применим изученные концепции для анализа финансового портфеля.

```python
# Практический пример: Анализ портфеля акций
np.random.seed(42)

# Симуляция доходностей двух акций
n_days = 252  # Торговых дней в году

# Акция A: высокая доходность, высокий риск
returns_A = np.random.normal(0.001, 0.02, n_days)  # 0.1% в день, σ=2%

# Акция B: умеренная доходность, умеренный риск, коррелированная с A
returns_B = 0.5 * returns_A + np.random.normal(0.0008, 0.015, n_days)

# Вычисление статистик
mean_A = np.mean(returns_A) * 252  # Годовая доходность
mean_B = np.mean(returns_B) * 252
std_A = np.std(returns_A) * np.sqrt(252)  # Годовая волатильность
std_B = np.std(returns_B) * np.sqrt(252)
cov_AB = np.cov(returns_A, returns_B)[0, 1] * 252  # Годовая ковариация
corr_AB = np.corrcoef(returns_A, returns_B)[0, 1]

print('Анализ портфеля акций')
print('=' * 60)
print(f'Акция A:')
print(f'  Годовая доходность: {mean_A:.2%}')
print(f'  Годовая волатильность: {std_A:.2%}')
print(f'\nАкция B:')
print(f'  Годовая доходность: {mean_B:.2%}')
print(f'  Годовая волатильность: {std_B:.2%}')
print(f'\nСвязь между акциями:')
print(f'  Ковариация: {cov_AB:.6f}')
print(f'  Корреляция: {corr_AB:.4f}')

# Портфель: 50% в A, 50% в B
w_A, w_B = 0.5, 0.5
returns_portfolio = w_A * returns_A + w_B * returns_B

mean_portfolio = np.mean(returns_portfolio) * 252
std_portfolio = np.std(returns_portfolio) * np.sqrt(252)

print(f'\nПортфель (50% A + 50% B):')
print(f'  Годовая доходность: {mean_portfolio:.2%}')
print(f'  Годовая волатильность: {std_portfolio:.2%}')

# Визуализация
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Кумулятивные доходности
cum_returns_A = np.cumprod(1 + returns_A)
cum_returns_B = np.cumprod(1 + returns_B)
cum_returns_portfolio = np.cumprod(1 + returns_portfolio)

axes[0, 0].plot(cum_returns_A, label='Акция A', color='blue')
axes[0, 0].plot(cum_returns_B, label='Акция B', color='red')
axes[0, 0].plot(cum_returns_portfolio, label='Портфель', color='green', linewidth=2)
axes[0, 0].set_xlabel('День')
axes[0, 0].set_ylabel('Кумулятивная доходность')
axes[0, 0].set_title('Динамика портфеля')
axes[0, 0].legend()

# Диаграмма рассеяния
axes[0, 1].scatter(returns_A, returns_B, alpha=0.3, color='purple')
axes[0, 1].set_xlabel('Доходность A')
axes[0, 1].set_ylabel('Доходность B')
axes[0, 1].set_title(f'Корреляция: {corr_AB:.3f}')

# Распределение доходностей
axes[1, 0].hist(returns_A, bins=50, alpha=0.5, label='A', density=True, color='blue')
axes[1, 0].hist(returns_B, bins=50, alpha=0.5, label='B', density=True, color='red')
axes[1, 0].hist(returns_portfolio, bins=50, alpha=0.5, label='Портфель', density=True, color='green')
axes[1, 0].set_xlabel('Дневная доходность')
axes[1, 0].set_ylabel('Плотность')
axes[1, 0].set_title('Распределение доходностей')
axes[1, 0].legend()
# Box plot
bp = axes[1, 1].boxplot([returns_A, returns_B, returns_portfolio])
axes[1, 1].set_xticklabels(['A', 'B', 'Портфель'])
axes[1, 1].set_ylabel('Дневная доходность')
axes[1, 1].set_title('Box plot сравнение')
plt.tight_layout()
plt.show()

# Эффект диверсификации
print(f'\nЭффект диверсификации:')
print(f'  Волатильность портфеля: {std_portfolio:.2%}')
print(f'  Средневзвешенная волатильность: {w_A*std_A + w_B*std_B:.2%}')
print(f'  Снижение риска: {(1 - std_portfolio/(w_A*std_A + w_B*std_B)):.2%}')
```
